```python
imxs = [0, 3, 7]
ixss = [0, 4, 9, 14, 19]

rescale = 4.

idm_dir = {}
IPk = {}

their = {}
their_ks = {}
their_pk = {}

for imx in imxs:
    
    mx = masses[imx]
    
    for ixs in ixss:
        
        sigma = sigmas[ixs]      
        
        idm_dir[mx, sigma, 1.] = Class()
        idm_dir[mx, sigma, 1.].set(base)
        idm_dir[mx, sigma, 1.].set(others)
        idm_dir[mx, sigma, 1.].set(verbose)
        idm_dir[mx, sigma, 1.].set(idm)
        idm_dir[mx, sigma, 1.].set(ppr)
        idm_dir[mx, sigma, 1.].set(flags)
        idm_dir[mx, sigma, 1.].set(sync)
        idm_dir[mx, sigma, 1.].set(proton)
        idm_dir[mx, sigma, 1.].set({'pow_n':0., 'sigma_0':1.*sigma, 'mx':mx})
        idm_dir[mx, sigma, 1.].set(extra)
        idm_dir[mx, sigma, 1.].compute()
        
        idm_dir[mx, sigma, rescale] = Class()
        idm_dir[mx, sigma, rescale].set(base)
        idm_dir[mx, sigma, rescale].set(others)
        idm_dir[mx, sigma, rescale].set(verbose)
        idm_dir[mx, sigma, rescale].set(idm)
        idm_dir[mx, sigma, rescale].set(ppr)
        idm_dir[mx, sigma, rescale].set(flags)
        idm_dir[mx, sigma, rescale].set(sync)
        idm_dir[mx, sigma, rescale].set(proton)
        idm_dir[mx, sigma, rescale].set({'pow_n':0., 'sigma_0':rescale*sigma, 'mx':mx})
        idm_dir[mx, sigma, rescale].set(extra)
        idm_dir[mx, sigma, rescale].compute()
        
        their[mx, sigma] = np.loadtxt(pk_root + 'def/n0_m{}_s{}_z0_pk.dat'.format(mx*1.e-3,sigma))
        their_ks[mx, sigma] = their[mx, sigma][:,0]
        their_pk[mx, sigma] = their[mx, sigma][:,1]
        
        kk = their_lcdm_ks # Nadler's values for k
        IPk[mx, sigma, 1.] = []
        IPk[mx, sigma, rescale] = []
        
        for k in kk:
            IPk[mx, sigma, 1.].append(idm_dir[mx, sigma, 1.].pk_lin(k*h, 0.)*h**3)
            IPk[mx, sigma, rescale].append(idm_dir[mx, sigma, rescale].pk_lin(k*h, 0.)*h**3)
        
        IPk[mx, sigma, 1.]=np.array(IPk[mx, sigma, 1.])
        IPk[mx, sigma, rescale]=np.array(IPk[mx, sigma, rescale])
```
